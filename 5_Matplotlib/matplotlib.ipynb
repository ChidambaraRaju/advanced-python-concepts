```python
t = np.arange(0., 5., 0.2)

plt.figure()

plt.subplot(3, 1, 1)
plt.plot(t, t, 'r--')
plt.xlabel('t values')
plt.ylabel('output values')
plt.title('Linear Subplot')

plt.subplot(3, 1, 2)
plt.plot(t, t**2, 'bs')
plt.xlabel('t values')
plt.ylabel('output values')
plt.title('Quadratic Subplot')

plt.subplot(3, 1, 3)
plt.plot(t, t**3, 'g^')
plt.xlabel('t values')
plt.ylabel('output values')
plt.title('Cubic Subplot')

plt.suptitle('Different functions of t')
plt.tight_layout()
plt.xlabel('t values')
plt.ylabel('output values')

plt.show()
```

```python
# axes - x axis, y axis, labels, tickers

x = np.linspace(1, 10, 100)

fig, axes = plt.subplots(2, 3)

axes[0,0].plot(x, np.sin(x))
axes[0,0].set_title('Sin func')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')

axes[0,1].plot(x, np.tan(x))
axes[0,1].set_title('Tan func')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')

axes[0,2].plot(x, np.exp(x))
axes[0,2].set_title('Exp func')
axes[0,2].set_xlabel('x')
axes[0,2].set_ylabel('y')

axes[1,0].plot(x, np.sqrt(x))
axes[1,0].set_title('Sqrt func')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')

axes[1,1].plot(x, np.log(x))
axes[1,1].set_title('Log func')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')

axes[1,2].plot(x, x)
axes[1,2].set_title('Linear func')
axes[1,2].set_xlabel('x')
axes[1,2].set_ylabel('y')


plt.tight_layout()
```

```python

x = np.random.rand(100) * 10
y = np.random.rand(100) * 10

plt.scatter(x,y, marker='D', edgecolors='black', linewidths=2, alpha=0.3, s = 50, label= 'random')

plt.legend(loc="lower right")
plt.grid(True)

plt.xlabel('x')
plt.ylabel('y')
plt.title('random scattered plot')
plt.xticks([0,5,10])
plt.yticks([0,5,10])

plt.show()
```
